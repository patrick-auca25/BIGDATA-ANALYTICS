Check current data in database

```python
query = "SELECT COUNT(*) as total_rows FROM energy_readings"
df = pd.read_sql(query, conn)

print(f" Current rows in database: {df['total_rows'][0]:,}")

if df['total_rows'][0] > 0:
    # Check date range
    date_query = """
        SELECT 
            MIN(timestamp)::timestamp as first_reading,
            MAX(timestamp)::timestamp as last_reading,
            COUNT(DISTINCT meter_id) as unique_meters
        FROM energy_readings
    """
    df_dates = pd.read_sql(date_query, conn)
    print(f"\n Data Range:")
    print(df_dates.to_string(index=False))
else:
    print("\nTable is empty - ready for fresh data")
```

```python
query = """
SELECT 
    COUNT(*) as total_rows,
    COUNT(DISTINCT meter_id) as unique_meters,
    MIN(timestamp)::timestamp as first_reading,
    MAX(timestamp)::timestamp as last_reading,
    (MAX(timestamp) - MIN(timestamp)) as duration
FROM energy_readings
"""

df = pd.read_sql(query, conn)

print("Current Status:")
print("-" * 60)
print(df.to_string(index=False))
print()

if df['total_rows'][0] > 0:
    cycles = df['total_rows'][0] // 500
    print(f" Cycles completed: {cycles}")
    print(f" Expected after 1 hour: 12 cycles (6,000 rows)")
    
    # Progress percentage (assuming 1-hour test)
    progress = (df['total_rows'][0] / 6000) * 100
    print(f" Progress: {progress:.1f}% of 1-hour test")
```
